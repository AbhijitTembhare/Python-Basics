2. What is an interpreter in Python?
- An interpreter is like a translator. It reads your Python code line by line and runs it.
3. What are pre-defined keywords in Python?
- Keywords are reserved words in Python that have special meanings. You can't use
them for anything else, like variable names. Examples: if, else, for, while.
4. Can keywords be used as variable names?
- No, you can't use keywords as variable names. Python will give you an error.
5. What is mutability in Python?
- Mutability means whether you can change something aŌer it's created. Some things
in Python can be changed (mutable), and some can't (immutable).
6. Why are lists mutable, but tuples are immutable?
- Lists are designed to be flexible. You oŌen need to add, remove, or change items in a
list. Tuples are designed to be fixed, like a record. Once you create a tuple, you can't
change it. This makes them a little faster and safer in some situaƟons.
7. What is the difference between “==” and “is” operators in Python?
- “==”checks if two things have the same value. is checks if two things are the same
object in memory (meaning they are literally the same thing, not just equal).
8. What are logical operators in Python?
- Logical operators let you combine or change true/false values. They are and, or, and
not.
9. What is type casting in Python?
- Type casting is when you change the data type of something. For example, turning a
number into a string or a string into a number.
10. What is the difference between implicit and explicit type casting?
- Implicit type casting is when Python does the type conversion automatically. Explicit
type casting is when you tell Python exactly what type you want using functions like
int(), str(), float().
11. What is the purpose of conditional statements in Python?
- Conditional statements (like if, elif, else) let your code make decisions. They run
different blocks of code depending on whether a condition is true or false.
12. How does the elif statement work?
- elif is short for "else if." It lets you check multiple conditions in a row. If the first if is
false, it checks the next elif, and so on.
13. What is the difference between for and while loops?
- A for loop is usually used when you know how many times you want to repeat
something. A while loop is used when you want to repeat something until a certain
condition becomes false.
14. Describe a scenario where a while loop is more suitable than a for loop.
- A while loop is better when you don't know in advance how many times you need to
loop. For example, if you're reading data from a file until you reach the end, or
waiting for a user to type a specific input.
